# Data Preparation (Data  Wrangling)

Import dataset

```r
pendudukjkt2020 <- read.csv('Penduduk DKI Jakarta tahun 2020.csv')
```

Show dataframe

```r
head(pendudukjkt2020)
```

This dataset is data of DKI Jakarta population 2020

Dataset columns: 

**tahun** is year

**nama_provinsi** is province

**nama_kabupaten_kota** is district

**nama_kecamatan** is subdistrict

**nama_kelurahan** is village

**usia** is age group

**jenis_kelamin** is gender

**jumlah_penduduk** is population

Profile dataframe

```r
str(pendudukjkt2020)
```

```r
summary(pendudukjkt2020)
```

See dataset columns

```r
names(pendudukjkt2020)
```

Check unique values of categorical columns

```r
print(unique(pendudukjkt2020$tahun))
print(unique(pendudukjkt2020$nama_provinsi))
print(unique(pendudukjkt2020$nama_kabupaten_kota))
print(unique(pendudukjkt2020$usia))
print(unique(pendudukjkt2020$jenis_kelamin))
```

Show total total population of DKI Jakarta in 2020

```r
totalpop = sum(pendudukjkt2020$jumlah_penduduk)
print(paste('Jakartan Total Population in 2020 is', totalpop))
```

# Data Visualization

Import tidyverse library for visualizing dataset

```r
library(tidyverse)
```

Show districts distribution

```r
ggplot(data=pendudukjkt2020) + geom_bar(mapping = aes(x = nama_kabupaten_kota, color = nama_kabupaten_kota))
```

Show population distribution by population genders

```r
ggplot(data=pendudukjkt2020) + geom_histogram(mapping = aes(x = jumlah_penduduk, color = jenis_kelamin))
```

Show population distribution of each districts

```r
ggplot(data = pendudukjkt2020) + 
  geom_histogram(mapping = aes(x = jumlah_penduduk, fill = jenis_kelamin)) + 
  facet_wrap(~ nama_kabupaten_kota, nrow = 3)
```

Show gender distribution by age

```r
ggplot(pendudukjkt2020 , aes(x = jumlah_penduduk, fill = jenis_kelamin)) + geom_bar() + facet_wrap('usia')
```

# Data Cleaning

Rename dataframe columns

```r
names(pendudukjkt2020)[c(1:8)] <- c('Tahun', 'Provinsi', 'Kabupaten', 'Kecamatan', 'Kelurahan', 'Kelompok Usia', 'Jenis Kelamin', 'Jumlah Penduduk')
head(pendudukjkt2020)
```

Delete the last column (X) by column index number

```r
pendudukjkt2020 <- pendudukjkt2020[-9]
head(pendudukjkt2020)
```

Check missing value

```r
sum(is.na(pendudukjkt2020))
```
